```python
import numpy as np 
import pandas as pd 

path = 'C:/Users/priye/OneDrive/Desktop/DataPer/MScDataScience/InternshipVideoAnalytics/Titanic data/'

df_train = pd.read_csv(path + "train.csv")
df_test = pd.read_csv(path + "test.csv")
```

```python
df_train
```

```python
#for training dataset

df_train = df_train.drop(columns= ['PassengerId','Name','Age','Ticket','Fare','Cabin','Embarked'])
df_train['Sex'].replace({'female':0,'male':1},inplace=True)
df_train

train_data = df_train.values

X = train_data[:,1:5]
Y = train_data[:,0]
```

```python
train_data
```

```python
from sklearn import preprocessing

min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)
```

```python
X_scale
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, Y_train, Y_val = train_test_split(X_scale, Y, test_size=0.3)
```

```python
print(X_train.shape, X_val.shape, Y_train.shape, Y_val.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential([Dense(32, activation='relu', input_shape=(4,)),    
                    Dense(32, activation='relu'),    
                    Dense(1, activation='sigmoid'),])
model.compile(optimizer='sgd',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, Y_train,
          batch_size=32, epochs=100,
          validation_data=(X_val, Y_val))
```

```python
model.evaluate(X_val, Y_val)[1]
```
